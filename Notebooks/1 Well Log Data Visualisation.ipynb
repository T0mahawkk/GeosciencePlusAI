#### I$\,\,\,\,$Visualise Lithology Distribution

<p> We start off with visualising the lithology distribution of the well to understand its geological composition.</p>

<p> We import the <code>VisualiseWellData</code> class from the <code>Lib.DataVisualisation</code> module and create an instance of it named <code>visualiser</code>. This class contains methods designed to visualise well data, including plotting lithology distributions, well logs, and crossplots from CSV files. The instance <code>visualiser</code> can then be used to call these methods and generate various visualisations to analyse and interpret well data more effectively.</p>

```python
from Lib.DataVisualisation import VisualiseWellData
visualiser = VisualiseWellData()

well_data_path = r".\Well Data\15_9-F-14.csv"
well_name = "15_9-F-14"
```

<p> The <code>visualise_lithology_distribution</code> function uses the well data from <code>15_9-F-14.csv</code> to create a bar plot that shows the count of each lithology type, with distinct colors and hatch patterns for each lithology.</p>

```python
visualiser.visualise_lithology_distribution(well_data_path, well_name)
```

#### II$\,\,\,\,$Explore Available Well Logs

<p> The <code>show_available_logs</code> function loads the well data (from the specified CSV file) and provides an overview of the available log data (i.e., the columns) in the file. It will display basic statistics for each column, including:</p>
<ul>
<li>Count of non-null entries</li>
<li>Mean and standard deviation of entries</li>
<li>Minimum and maximum entries</li>
</ul>
<p>This helps to quickly assess the quality and range of the available log data, making it easier to identify any issues, like missing data.</p>

```python
visualiser.show_available_logs(well_data_path)
```

#### III$\,\,\,\,$Visualise Well Logs

<p>We can generate well log plots using the <code>plot_well_logs_and_lithology</code> function to visualise various logs alongside the vertical lithology profile. This type of plot, often referred to as a <em>well log composite</em> plot, helps to display multiple well log parameters and the lithological changes over depth in a clear, side-by-side manner.</p>

```python
visualiser.plot_well_logs_and_lithology(well_data_path, well_name)
```

#### IV$\,\,\,\,$Create 2D Crossplots

We can also create a 2D crossplot of any two variables - in this case, NPHI vs. RHOB, using the <code>crossplot_2D</code> function to examine the relationship between them.

```python
visualiser.crossplot_2D(well_data_path, well_name, "NPHI", "RHOB")
```
